3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
high_level = 5
s_high = 140
d_high = 90

hsl = df[df["Stress Levels"] > high_level][["Blood Pressure (s/d)", "Stress Levels"]].sort_values(by="Stress Levels", ascending=False)

bp = df["Blood Pressure (s/d)"].str.split("/", expand=True).astype(int)
bp.columns = ["s", "d"]
bp = bp[(bp["s"] > s_high) | (bp["d"] > d_high)]

matching_indx = hsl.index.intersection(bp.index)

count_with_high_stress = len(hsl)
count_with_high_bp  = len(bp)
count_with_high_bp_and_stress  = len(matching_indx)

print(f"Кол-во people с высоким давлением: {count_with_high_bp}")
print(f"Кол-во people с высоким стрессом: {count_with_high_stress}")
print(f"Кол-во people с высоким стрессом и давлением: {count_with_high_bp_and_stress}")

high_stress_no_bp = count_with_high_stress - count_with_high_bp_and_stress

high_bp_no_stress = count_with_high_bp - count_with_high_bp_and_stress

print()
print(f"Кол-во people с высоким стрессом, но нормальным давлением: {high_stress_no_bp}")
print(f"Кол-во people с высоким давлением, но нормальным стрессом: {high_bp_no_stress}")
print()

if count_with_high_bp_and_stress > high_stress_no_bp and count_with_high_bp_and_stress > high_bp_no_stress:
    print("Да, люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу")
else:
    print("Нет четкой зависимости между высоким уровнем стресса и высоким давлением")
````

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

````python
import numpy as np

bmi_30 = np.percentile(df["BMI"], 30)
bmi_60 = np.percentile(df["BMI"], 60)
bmi_75 = np.percentile(df["BMI"], 75)

high_bmi = df[df["BMI"] > bmi_75][["Chronic Diseases"]].notna().mean()
mid_bmi = df[(df["BMI"] >= bmi_30) & (df["BMI"] <= bmi_60)][["Chronic Diseases"]].notna().mean()

print(high_bmi)
print(mid_bmi)

if high_bmi.get("Chronic Diseases") > mid_bmi.values[0]:
    print("Правда")
else:
    print("Неправда")
````

````python
# У нас три параметра, один из низ категориальный, поэтому я решил использовать пузырьковую диаграмму, где цвет пузыря будет уровень дохода

income_map = {"Low": 1, "Medium": 2, "High": 3}
inf = df[["Cognitive Function", "Stress Levels", "Income Level"]].copy()
inf["Income num"] = inf["Income Level"].map(income_map)

plt.figure(figsize=(10, 10))

sns.scatterplot(
    x=inf["Stress Levels"],
    y=inf["Cognitive Function"],
    hue=inf["Income Level"], 
    size=inf["Income num"],  
    palette="coolwarm",      
    sizes=(10, 300),
    alpha=0.7
)

plt.title("Связь между стрессом, когнитивными способностями и уровнем дохода")
plt.xlabel("Уровень стресса")
plt.ylabel("Когнитивные способности")
plt.legend(title="Уровень дохода", bbox_to_anchor=(1, 1))

plt.show()
````
